```python
import requests
from bs4 import BeautifulSoup
import nbformat
from nbformat.v4 import new_notebook, new_markdown_cell, new_code_cell


def loadContent(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    return soup
    
def gfg2Notebook(url, filename):
    soup = loadContent(url)
    
    cells = []
    
    # Extract page title
    title_tag = soup.find("h1")
    header = "# " + title_tag.get_text(strip=True)
    cells.append(new_markdown_cell(header))

    # Extract page body
    content = f"## Description\n- [Problem]({url})\n"

    body = soup.find("div",class_="text")
    description = body.find("p").find("span").get_text(strip=True)
    content += f"- {description}\n"
    content += f"{body.find("blockquote").find("p")}"
    cells.append(new_markdown_cell(content))
    cells.append(new_code_cell(''))
    
    nb = new_notebook(cells=cells)
    with open(filename, "w", encoding="utf-8") as f:
        nbformat.write(nb, f)
    print(f"Notebook saved as '{filename}'")
    return title_tag.get_text(strip=True)
```

```python
def tag(x,places=2):
    return f"{x}".rjust(places, '0')

def filePath(x,y,root):
    return f"{root}/{tag(x,2)}.{tag(y,3)}.ipynb"

def alreadyExists(filePath):
    try:
        with open(filePath,'r'):
            return True
    except:
        pass
    return False

def updateIndex(title,file_path,index):
    path = f"/lab/tree/{file_path}".replace('./','')
    print(path)
    uuid = file_path.split('/')[-1].rsplit('.',1)[0]
    return index + f"- [ ] [{uuid} {title}]({path})\n"
    
def generate(urls,config):
    x = config['X']
    start = config['Y']
    end = start + len(urls)
    root = config['ROOT']
    index = ""
    
    for y in range(start,end):
        file_path = filePath(x,y,root)
        url = urls[y-start]
        print(f"{y-start+1}/{end-start}",file_path,end=" ")
        exists = alreadyExists(file_path)
        if exists:
            print("Already Exists, Skipping...")
        else:
            print("Creating...")
            title = "HERER"
            try:
                title = gfg2Notebook(url,file_path)
                index = updateIndex(title,file_path,index)
            except Exception as e:
                print("Error", e)
                break
    print("---- Completed -----")
    print("\n--- Index Here -----\n")
    print(index)
```
